# Seaborn scatter plots

```python
sns.scatterplot(x = 'total_bill',y = 'tip', data = tips)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "size")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6, 2))

sns.histplot(data = tips, x = "total_bill", hue = "sex", multiple = "stack")

plt.title("Total Bill Distribution By Gender")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create the histogram
sns.histplot(data=tips, x='total_bill', hue='sex', multiple='stack')

# Add title and labels
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create the histogram
sns.histplot(data=tips, x='total_bill', hue='sex', multiple='stack')

# Add title and labels
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create the histogram
sns.histplot(data=tips, x='total_bill', hue='sex', multiple='dodge')

# Add title and labels
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
tips["day_time"] = tips["day"].astype(str) + "-" + tips["time"].astype(str)

sns.histplot(data=tips, x="total_bill",hue="day_time",multiple="stack",shrink=1)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill",hue="day_time",multiple="dodge",shrink=1)

plt.xlabel("Total Bill")
plt.show()
```

```python
sns.histplot(data=tips,x="total_bill",hue="day",legend=True,multiple="dodge")
```
